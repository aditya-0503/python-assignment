```python
# 3.Using the sales data, generate a heatmap to visualize the variation in sales across
# different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

fig = px.imshow(pivot_table,
                color_continuous_scale='Viridis',
                labels=dict(x='Month', y='Day', color='Sales'),
                title='Sales Heatmap by Month and Day')

fig.update_xaxes(title_text='Month')
fig.update_yaxes(title_text='Day')
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(^(X^2+y^2))
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x[0], y=y[:, 0])])
fig.update_layout(title='Surface Plot', scene=dict(zaxis=dict(range=[-1, 1])))
fig.show()
```

**BOKEH ASSIGNMENT:**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.tile_sources import get_tile
```

```python
from bokeh.plotting import figure, output_file, show
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
output_file('sinx')

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
```

```python
output_file('X and Y plot')

np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.randint(5, 50, size=n)
colors = np.random.choice(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], size=n)

source = ColumnDataSource(data=dict(x=x, y=y, size=sizes, color=colors))

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter('x', 'y', size='size', color='color', legend_field='color', source=source)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
output_file('fruits_count')

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=800, height=400)

p.vbar(x='fruits', top='counts', width=0.6, source=source,
       line_color='white', fill_color='blue')

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
output_file('data_hist')
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value', y_axis_label='Count',
           width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='blue', line_color='black', alpha=0.7)

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
output_file('dataset')

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}

p = figure(title="Heatmap",
           x_axis_label='X', y_axis_label='Y',
           width=400, height=400)

mapper = linear_cmap('value', palette='Viridis256', low=min(data['value']), high=max(data['value']))

p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
        line_color=None, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
