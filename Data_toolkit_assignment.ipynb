3. Explain the following terms:

 The difference in np.array, np.asarray and np.asanyarray

 The difference between Deep copy and shallow copy.

```python
# np.array = This function creates a new NumPy array from any object exposing the array interface, including lists, tuples, or other NumPy arrays.
# np.asarray = This function is used to convert input data into a NumPy array, but unlike np.array(), it does not copy the data if the input is already a NumPy array.
# np.asanarray = This function is used to convert input data into a NumPy array, but unlike np.array(), it does not copy the data if the input is already a NumPy array.
```

```python
# Shallow copy >> A shallow copy creates a new object but does not recursively copy the objects contained within the original object. Instead, it copies references to the objects
# found inside the original object.

# Deep copy >> A deep copy creates a new object and recursively copies all objects contained within the original object. This means that the deep copy will have its own copy of all
# nested objects, and changes to the original nested objects will not affect the deep copy.
```

14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(15, 30, size=100)
humidity = np.random.uniform(40, 90, size=100)
data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(data['Date'], data['Temperature'], color='tab:red', label='Temperature', linewidth=2)
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='tab:blue', label='Humidity', linewidth=2)
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
```

```python
x = np.random.rand(100) * 2 - 1
y = np.random.rand(100) * 2 - 1

data = pd.DataFrame({'x': x, 'y': y})

def get_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Quadrant I (Q1)'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Quadrant II (Q2)'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III (Q3)'
    else:
        return 'Quadrant IV (Q4)'

data['Quadrant'] = data.apply(get_quadrant, axis=1)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set1', s=100, edgecolor='black')

plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Quadrant-wise Scatter Plot')

plt.legend(title='Quadrant', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show, output_file

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

p.grid.grid_line_alpha = 0.3
p.grid.grid_line_width = 1
p.grid.grid_line_color = "black"

show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
from bokeh.models import ColumnDataSource, HoverTool

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

data = pd.DataFrame({
    'category': categories,
    'value': values
})

source = ColumnDataSource(data)

p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

bars = p.vbar(x='category', top='value', source=source, width=0.6, color='skyblue', legend_field="category")

hover = HoverTool()
hover.tooltips = [("Category", "@category"), ("Value", "@value")]
p.add_tools(hover)

p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

p.grid.grid_line_alpha = 0.3

p.legend.title = "Categories"
p.legend.location = "top_left"

show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
